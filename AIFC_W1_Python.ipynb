#**3. Function Construction**
### General structure of a function

```python
# example
def compute_rectangle_area (height, width):
    '''
    This function aims to compute area for a rectangle.
    height -- the height of the rectangle
    width -- the width of the rectangle
    This function returns the area of the rectangle
    
    '''
    area = height*width
    return area
area1 = compute_rectangle_area(5,6)
print('area 1:', area1)

area2 = compute_rectangle_area(height=5, width=6)
print('area 2:',area2)

area3 = compute_rectangle_area(0,0)#defaut value height=0, width=0
print('area 3:',area3)
```

```python
# example
def ReLU(x):
    '''
    This function aims to compute ReLU for a value x.
    x -- an input value
    This function returns the ReLU of x
    :param x:
    :return:
    '''
    result = 0
    if x > 0:
        result = x
    return result
value1 = ReLU(x=5)
value2 = ReLU(x=-2)

print(value1)
print(value2)
```

```python
# iterate a list
fruits = ['apple','banana', 'melon', 'peach']
for fruit in fruits:
  print(fruit)
```



> **break keyword**



```python
# duyet phan tu trong range(10)
for i in range (10):
    #hoi phan tu thu i co bang 5 khong?
    if i == 5:
      #neu bang thi thoat vong lap for nay
      break
    #lam gi do voi i
    print('Gia tri i la',i)
```



> **continue keyword**



```python
# duyet phan tu trong range(10)
for i in range(10):
  # hoi phan tu i co bang 5 khong?
  if i == 5:
    # neu bang thi goi continue
    # phan code sau continue se khong duoc thuc hien trong phan lap nay
    continue
  # lam gi do voi i
  print('Gia tri cua i la',i)
```
