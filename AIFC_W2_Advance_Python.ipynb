```python
'''
Euler's number
'''
# dinh nghia ham tinh giai thua
def factorial(n):
  result = 1 # dau tien gan cho bien result =1

  for i in range(2, n+1): #range(3,6) se tra ve ket qua la 3,4,5
      result = result*i

  return result

# dinh nghia ham uoc luong so e
def estimate_e(n):
    result = 1

    for i in range(1, n+1):
        result = result + 1/ factorial(i)

    return result





```

```python
# uoc luong so e voi n = 30
estimate_e(30)
```

```python
'''
Compute quadratic root for the number N
'''
def compute_square_root(N, num_loops):
  '''
  This function aims to compute square root for the number N

  N -- the number needs to take the square root
  num_loops --number of loops used for this optimization 
  '''
  x_n = N/2.0

  for i in range(num_loops):
    x_np1 = (x_n + N/x_n)/2.0
    x_n = x_np1

  return x_np1

print(compute_square_root(N=9, num_loops=10))  
print(compute_square_root(N=7, num_loops=10))  
```
