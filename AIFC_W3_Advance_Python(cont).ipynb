Example: Solve quadratic equation

```python
import math
def quadratic_equation(a,b,c):
  '''
  This function aims at solving the quadratic equation a,b,c---three parameters and a =! 0
  '''
  #compute delta
  delta = b*b -4*a*c
  if delta < 0:
    return ()
  elif delta == 0:
    x = -b/(2*a)
    return (x,)
  else:
    x1 = (-b+math.sqrt(delta))/(2*a)
    x2 = (-b-math.sqrt(delta))/(2*a)
    return (x1, x2)
```

```python
result = quadratic_equation(a=5,b=0,c=1)
print(type(result))
print(len(result))
print(result)
```

```python
result = quadratic_equation(a=5,b=5,c=1)
print(type(result))
print(len(result))
print(result)
```

```python
result = quadratic_equation(a=4,b=4,c=1)
print(type(result))
print(len(result))
print(result)
```

# **3. Dictionary**
Structure:

dictionary_name = {key-1:value-1,...,key-n:value-n}

Creata a dictionary

```python
# Create a dictionary
parameters = {'learning_rate': 0.1,
              'optimizer':'Adam',
              'metric':'Accuracy'}
print(parameters)
print(type(parameters))
```

Update a value

```python
parameters = {'learning_rate': 0.1,
              'metric':'Accuracy'}
parameters['learning_rate'] = 0.2
print(parameters)
print(type(parameters))
```

Copy a dictionary

```python
parameters = {'learning_rate': 0.1,
              'metric':'Accuracy'}
a_copy = parameters.copy()             
a_copy['learning_rate'] = 0.2
print(parameters)
print(a_copy)
```

Ham copy() chi sao chep kieu shallow

```python
d1 = {'a':[1,2], 'b':5}
d2 = d1.copy()

#thay doi gia tri d2 se anh huong den d1
d2['a'][0] = 3
d2['a'][1] = 4

print('d1:',d1)
print('d2:',d2)
```

Su dung ham deepcopy() trong module copy

```python
import copy

d1 = {'a':[1,2], 'b':5}
d2 = copy.deepcopy(d1)

# thay doi gia tri d2
d2['a'][0] = 3
d2['a'][1] = 4


print('d1:',d1)
print('d2:',d2)
```

```python
parameters = {'learning_rate': 0.1,
              'optimizer':'Adam',
              'metric':'Accuracy'}

for key in parameters:
  print(key)
```

```python
parameters = {'learning_rate': 0.1,
              'optimizer':'Adam',
              'metric':'Accuracy'}
items = parameters.items()
for key, value in items:
  print(key, value)
```

Get a value by a key

```python
# Get value using get() function
parameters = {'learning_rate': 0.1,
              'optimizer':'Adam',
              'metric':'Accuracy'}
value = parameters.get('learning_rate')
print(value)

print('\nAfter using get() function')
print(parameters)
```

```python
# Get value and delete the corresponding item
parameters = {'learning_rate': 0.1,
              'optimizer':'Adam',
              'metric':'Accuracy'}
value = parameters.pop('learning_rate')
print(value)

print('\nAfter using pop() function')
print(parameters)
```

popitem(): lay ra mot phan tu o cuoi dictionary

```python

parameters = {'learning_rate': 0.1,
              'optimizer':'Adam',
              'metric':'Accuracy'}
item = parameters.popitem()

print(item)
print(parameters)
```

clear(): xoa tat ca cac phan tu cua mot dictionary

```python
parameters = {'learning_rate': 0.1,
              'metric':'Accuracy'}
print('Before using clear() function')
print(parameters)

parameters.clear()

print('\nAfter using clear() function')
print(parameters)
```

Use del keyword to delete an item
